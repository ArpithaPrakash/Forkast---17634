```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

sample_data = pd.read_csv('zomato.csv')

sample_data['approx_cost(for two people)'] = (
    sample_data['approx_cost(for two people)']
    .str.replace(',', '')  # Remove commas
    .astype(float)  # Convert to float
)

sample_data['rate'] = sample_data['rate'].astype(str)
sample_data = sample_data.dropna(subset=['rate']) 
sample_data = sample_data[~sample_data['rate'].isin(['NEW', '-'])] 
sample_data['rate'] = sample_data['rate'].str.strip().str.split('/').str[0].astype(float)

sample_data['rest_type'] = sample_data['rest_type'].str.split(',')
sample_data = sample_data.explode('rest_type')
sample_data['rest_type'] = sample_data['rest_type'].str.strip()

sample_data['online_order'] = sample_data['online_order'].map({'Yes': 1, 'No': 0})
sample_data['book_table'] = sample_data['book_table'].map({'Yes': 1, 'No': 0})

sample_data['cuisines'] = sample_data['cuisines'].str.split(',')
sample_data = sample_data.explode('cuisines')
sample_data['cuisines'] = sample_data['cuisines'].str.strip()


sample_data = sample_data.drop(columns=['url', 'address', 'phone', 'menu_item', 'name', 'reviews_list', 'dish_liked'])
sample_data.dropna(inplace=True)
sample_data.columns
```

```python

sample_data = pd.get_dummies(sample_data, columns=['location', 'rest_type', 'cuisines', 'listed_in(type)', 'listed_in(city)'], drop_first=True)
# Splitting features and target
X = sample_data.drop(columns=['rate'])
y = sample_data['rate']

# Standardizing the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Splitting into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Feature names and number of features
features = X.columns.tolist()
num_features = len(features)

# Lists to store results
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

train_r2_scores = []
test_r2_scores = []
train_errors = []
test_errors = []

# Iteratively fit the model with increasing number of features
for i in range(1, num_features + 1):
    # Select the first i features
    X_train_subset = X_train[:, :i]
    X_test_subset = X_test[:, :i]

    # Fit the model
    model = LinearRegression()
    model.fit(X_train_subset, y_train)

    # Train and test predictions
    y_train_pred = model.predict(X_train_subset)
    y_test_pred = model.predict(X_test_subset)

    # Calculate accuracy metrics
    train_r2 = r2_score(y_train, y_train_pred)
    test_r2 = r2_score(y_test, y_test_pred)
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)

    # Append results
    train_r2_scores.append(train_r2)
    test_r2_scores.append(test_r2)
    train_errors.append(train_mse)
    test_errors.append(test_mse)

# Plotting the bias-variance tradeoff
import matplotlib.pyplot as plt
plt.figure(figsize=(32, 10))
plt.plot(range(1, num_features + 1), train_r2_scores, label="Train RÂ²", marker='o')
plt.plot(range(1, num_features + 1), test_r2_scores, label="Test RÂ²", marker='o')
plt.title("Model Complexity vs RÂ² (Bias-Variance Tradeoff)", fontsize=14)
plt.xlabel("Number of Features", fontsize=12)
plt.ylabel("RÂ² Score", fontsize=12)
plt.xticks(range(1, num_features + 1, max(1, num_features // 10)))
plt.grid()
plt.legend(fontsize=12)
plt.tight_layout()
plt.savefig('bias_variance.png', dpi=300, bbox_inches='tight')  # Save the plot
plt.show()
```
